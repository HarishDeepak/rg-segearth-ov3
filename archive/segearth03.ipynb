```python
import shutil
import numpy as np
from PIL import Image
from pathlib import Path

RAW_DIR = Path('/kaggle/input/datasets/rachanaamraghuthama/rskt-potsdam-test-data/rskt-weights/')
ANN_DIR = Path('/kaggle/working/potsdam_format/annotations_index/test/')
MINI_RAW_DIR = Path('/kaggle/working/mini_test/images/')
MINI_ANN_DIR = Path('/kaggle/working/mini_test/annotations/')

for d in [ANN_DIR, MINI_RAW_DIR, MINI_ANN_DIR]:
    d.mkdir(parents=True, exist_ok=True)

COLOR_MAP = {
    (255, 255, 255): 0,  # Road
    (0, 0, 255): 1,      # Building
    (0, 255, 255): 2,    # Grass
    (0, 255, 0): 3,      # Tree
    (255, 255, 0): 4,    # Car
    (255, 0, 0): 5,      # Clutter
    (0, 0, 0): 255       # Ignore
}

all_labels = sorted(list(RAW_DIR.glob('*_label_noBoundary.tif')))

if len(all_labels) == 0:
    print(f"ERROR: No files ending in '_label_noBoundary.tif' found in {RAW_DIR}")
    print("Please double check that this is the folder containing the images, not the model weights.")
else:
    for target_mask_path in all_labels:
        # Convert mask
        rgb = np.array(Image.open(target_mask_path).convert('RGB'))
        h, w, _ = rgb.shape
        indexed = np.full((h, w), 255, dtype=np.uint8)

        for color, cls_id in COLOR_MAP.items():
            mask = np.all(rgb == color, axis=-1)
            indexed[mask] = cls_id

        out_file = ANN_DIR / (target_mask_path.stem + '.png')
        Image.fromarray(indexed).save(out_file)

        # Copy to mini_test
        raw_name = target_mask_path.name.replace('_label_noBoundary.tif', '_RGB.tif')
        
        try:
            shutil.copy(RAW_DIR / raw_name, MINI_RAW_DIR / raw_name)
            shutil.copy(out_file, MINI_ANN_DIR / out_file.name)
            print(f"Successfully converted: {target_mask_path.name}")
        except FileNotFoundError:
            print(f"WARNING: Found mask {target_mask_path.name} but could not find matching RGB image {raw_name}")
            
    print("\nAll available files have been mapped and copied to the mini_test directories!")
```

```python
%%writefile /kaggle/working/SegEarth-OV-3/run_manual_eval.py
import os
import time
import torch
import numpy as np
from pathlib import Path
from PIL import Image
from sklearn.metrics import confusion_matrix

# Explicitly import the custom model
import segearthov3_segmentor 
from mmseg.apis import init_model, inference_model

Image.MAX_IMAGE_PIXELS = None

config_file = 'configs/cfg_potsdam.py' 

checkpoint_file = None 

IMG_DIR = Path('/kaggle/working/mini_test/images/')
ANN_DIR = Path('/kaggle/working/mini_test/annotations/')
OUTPUT_DIR = Path('/kaggle/working/manual_predictions')

# Create directories for our raw tiles, final indexed masks, and RGB visuals
TMP_TILE_DIR = OUTPUT_DIR / 'tmp_tiles'
MASK_DIR = OUTPUT_DIR / 'stitched_masks'
VIS_DIR = OUTPUT_DIR / 'stitched_visuals'

for d in [TMP_TILE_DIR, MASK_DIR, VIS_DIR]:
    d.mkdir(parents=True, exist_ok=True)

print("Initializing SAM3 Model on GPU...")
model = init_model(config_file, checkpoint_file, device='cuda:0')

classes = ['Road', 'Building', 'Grass', 'Tree', 'Car', 'Clutter']

palette = [
    [255, 255, 255],  # 0: Road (White)
    [0, 0, 255],      # 1: Building (Blue)
    [0, 255, 255],    # 2: Grass (Cyan)
    [0, 255, 0],      # 3: Tree (Green)
    [255, 255, 0],    # 4: Car (Yellow)
    [255, 0, 0]       # 5: Clutter (Red)
]

all_images = sorted(list(IMG_DIR.glob('*.tif')))
patch_size = 1024

# Initialize a global confusion matrix for all images
total_cm = np.zeros((6, 6), dtype=np.int64)

for img_path in all_images[2:4]:
    print(f"\n---> Processing {img_path.name} <---")
    img = Image.open(img_path)
    width, height = img.size
    
    # Create a blank canvas to hold the stitched 0-5 integer mask
    stitched_mask = np.zeros((height, width), dtype=np.uint8)
    
    start_time = time.time()
    tile_count = 0

    for y in range(0, height, patch_size):
        for x in range(0, width, patch_size):
            right = min(x + patch_size, width)
            bottom = min(y + patch_size, height)
            
            # Crop and save temporary tile (using PNG to avoid compression)
            patch = img.crop((x, y, right, bottom))
            tmp_path = TMP_TILE_DIR / 'tmp_patch.png'
            patch.save(tmp_path)
            
            # Run Inference
            result = inference_model(model, str(tmp_path))
            
            # Extract the raw 2D array of predictions (0s, 1s, 2s...)
            pred_mask = result.pred_sem_seg.data[0].cpu().numpy()
            
            # Paste the predicted array into our master canvas
            stitched_mask[y:bottom, x:right] = pred_mask[:(bottom-y), :(right-x)]
            
            del result
            torch.cuda.empty_cache()
            tile_count += 1

    elapsed = time.time() - start_time
    print(f"Segmented and stitched {tile_count} tiles in {elapsed:.1f}s.")
    
    # Save the raw indexed mask
    mask_out = MASK_DIR / img_path.name.replace('.tif', '_pred.png')
    Image.fromarray(stitched_mask).save(mask_out)
    
    # Create and save the clean RGB visual
    rgb_mask = np.zeros((height, width, 3), dtype=np.uint8)
    for i, color in enumerate(palette):
        rgb_mask[stitched_mask == i] = color
        
    vis_out = VIS_DIR / img_path.name.replace('.tif', '_vis.png')
    Image.fromarray(rgb_mask).save(vis_out)
    
    gt_path = ANN_DIR / img_path.name.replace('_RGB.tif', '_label_noBoundary.png')
    if gt_path.exists():
        gt_mask = np.array(Image.open(gt_path))
        
        # Flatten both masks
        gt_flat = gt_mask.flatten()
        pred_flat = stitched_mask.flatten()
        
        # Filter out the "Ignore" class (255) from the calculation
        valid_idx = (gt_flat != 255)
        gt_valid = gt_flat[valid_idx]
        pred_valid = pred_flat[valid_idx]
        
        # Add to our running confusion matrix
        cm = confusion_matrix(gt_valid, pred_valid, labels=[0, 1, 2, 3, 4, 5])
        total_cm += cm

print("\n===============================================")
print("             FINAL METRICS SUMMARY               ")
print("===============================================")

intersection = np.diag(total_cm)
ground_truth_set = total_cm.sum(axis=1)
predicted_set = total_cm.sum(axis=0)
union = ground_truth_set + predicted_set - intersection

ious = []
for i in range(6):
    iou = intersection[i] / union[i] if union[i] > 0 else float('nan')
    ious.append(iou)
    print(f"IoU for {classes[i]:<15}: {iou * 100:.2f}%")

# Slice the list to calculate mean IoU for indices 0 to 4 (ignoring 5: Clutter)
valid_ious = [ious[i] for i in range(5) if not np.isnan(ious[i])]
miou_no_clutter = np.mean(valid_ious)

print("-" * 47)
print(f"mIoU (Strictly Excluding Clutter): {miou_no_clutter * 100:.2f}%")
print("===============================================")
print(f"Clean RGB visuals saved to: {VIS_DIR}")
```
